```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
crashes = sns.load_dataset('car_crashes')
crashes.head()
```

```python
tips
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day');
```

# **Distribution plots in Seaborn**

**1.Distribution Plot**

```python
sns.distplot(crashes['not_distracted'], kde=False)
```

**Joint Plot**

```python
sns.jointplot(x='speeding', y='alcohol', data=crashes, kind='kde');
```

**KDE Plot**

```python
#KDE = Kernel Density Estimation
sns.kdeplot(crashes['alcohol']);
```

**Pair Plot**

```python
sns.pairplot(tips, hue='sex', palette='bright');
```

**Rug Plot**

```python
sns.rugplot(tips['tip']);
```

**Styling**

```python
sns.set_style('dark')

sns.set_context('paper', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crashes);
```

# Categorical Plots

## Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips);
```

## Count Plots

```python
#counts the occurence of a particular value. -Just like a histogram
sns.countplot(x='sex', data=tips);
```

## Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex');
```

## Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True);
```

## Strip Plot

```python
sns.set_style('white')

plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True);
```

## Palette

```python
#Check Matplotlib color map on internet to see different kinds of palettes
plt.figure(figsize=(8,5))

sns.set_style('dark')
sns.set_context('paper', font_scale=1.5)
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True, palette='magma');
```

# Matrix Plots

## Heat Maps

```python
plt.figure(figsize=(8,6))

crashes_mx = crashes.corr()
crashes_mx

sns.heatmap(crashes_mx, annot=True, cmap='Blues');
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
flights
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1);
```
